```python
import requests
import pandas
from lxml import etree
from time import sleep
from google.colab import drive
import os

headers = {
 "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.114 Safari/537.36"
}

main_url = "https://www.wassilykandinsky.net/"
url_data = requests.get(main_url, headers=headers).text
data = etree.HTML(url_data)
year_list = data.xpath('//*[@id="ad"]/div[2]/div[3]/a/@href')

os.makedirs("picture/Kandinsky")

for year in year_list:
  page_data = requests.get(main_url + year, headers=headers).text
  page_info = etree.HTML(page_data)
  img_url = page_info.xpath('//*[@id="main"]/div[2]/div/a/@href')
  year = page_info.xpath('//*[@id="main"]/h1/text()')

  if img_url:
    os.makedirs('picture/Kandinsky/' + str(year[0]))
    img_list = pandas.unique(img_url).tolist() #去除重覆資料並保留順序

    for img in img_list:
      img_data = requests.get(main_url + img, headers=headers).content
      img_info = etree.HTML(img_data)
      img_name = img_info.xpath('//*[@id="main"]/h1/text()')
      img_big_url = img_info.xpath('//*[@id="img_big"]/@src')
      img_pic = requests.get(main_url + str(img_big_url[0]), headers=headers).content

      try:
        with open('picture/Kandinsky/' + str(year[0]) + '/' + str(img_name[0]) + '.jpg', 'wb') as f:
          print('正在下載' + str(year[0]) + '年 - ' + str(img_name[0]))
          f.write(img_pic)
      except Exception as err: print(err)
      sleep(1)
  sleep(1)
```
